# Hodgkin-Huxley Model

$$
    C_m \frac{dV}{dt} = g_{\text{Na}} m^3 h (E_{\text{Na}} - V) + g_{\text{K}} n^4 (E_{\text{K}} - V) + g_\text{L} (E_{\text{L}} - V)
$$

$$
    \begin{align*}
        \frac{dn}{dt} &= \alpha_n(V)(1 - n) - \beta_n(V)n \\
        \frac{dm}{dt} &= \alpha_m(V)(1 - m) - \beta_m(V)m \\
        \frac{dh}{dt} &= \alpha_h(V)(1 - h) - \beta_h(V)h
    \end{align*}
$$

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

# Load data
df = pd.read_csv("hh_data.csv", sep="\t")

# Create subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 8))
plt.suptitle("Hodgkin-Huxley Model")

axs[0].plot(df['t'], df['V'], label='Action Potential (V)')
axs[0].set_ylabel("Voltage (mV)")
axs[0].set_title("Action Potential")
axs[0].legend()

axs[1].plot(df['t'], df['n'], label='n (K+ gate)',          color='dodgerblue')
axs[1].plot(df['t'], df['m'], label='m (Na+ activation)',   color='mediumseagreen')
axs[1].plot(df['t'], df['h'], label='h (Na+ inactivation)', color='orange')
axs[1].set_xlabel("Time (ms)")
axs[1].set_ylabel("Gating Value")
axs[1].set_title("Gating Variables")
axs[1].legend()

plt.tight_layout()
plt.show()
```
